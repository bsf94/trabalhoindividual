#***Projeto individual:* Análise da relação entre liberdade econômica e o crescimento do PIB no Brasil.**

- **OBJETIVO:** O objetivo deste projeto é analisar a relação entre o crescimento do PIB no Brasil e o nível de liberdade econômica, conforme medido pelo Índice Economic Freedom of the World (EFW). Além disso, busca-se identificar quais subcomponentes desse índice impactam mais significativamente o nível geral de liberdade econômica no Brasil, bem como identificar quais subcomponentes apresentam maior correlação com as variações no crescimento do PIB. O projeto também visa desenvolver um modelo preditivo para determinar qual subcomponente precisa ser alterado para permitir um maior crescimento do PIB.

- **RELEVÂNCIA:** Através dessa análise, espera-se identificar os subcomponentes específicos do Índice EFW que têm maior impacto no nível geral de liberdade econômica no país e compreender quais subcomponentes estão mais correlacionados com as variações no crescimento do PIB. Essas informações podem ser úteis para o entendimento de fatores relacionados à políticas públicas que visem melhorar a liberdade econômica e impulsionar o crescimento econômico sustentável no Brasil. Ao fornecer insights sobre quais aspectos precisam ser priorizados e quais ações podem ser tomadas, espera-se entender, a partir do índice EFW, quais fatores contribuem para o "atraso" econômico do Brasil.

- **CONTEXTO:** Este projeto tem como foco o Brasil e busca analisar a relação entre o crescimento do PIB e a liberdade econômica no país. Para isso, serão utilizados dados entre o período de 2000 a 2020. Esse intervalo de tempo permitirá uma análise ampla e aprofundada das tendências e correlações ao longo das últimas duas décadas.

- **LIMITAÇÕES:** Este projeto apresenta algumas limitações importantes a serem consideradas. Primeiramente, é fundamental reconhecer que correlação não implica causalidade. Mesmo que seja possível identificar correlações entre os indicadores de liberdade econômica e o crescimento do PIB, estabelecer uma relação causal é mais complexo. A liberdade econômica é apenas um dos muitos fatores que influenciam o crescimento do PIB, e há muitas outras variáveis que impactam o PIB de um país, como estabilidade política e social, investimento em infrastrutura, desenvolvimento de capital humano, nível de educação da população, condições macroeconômicas nacionais e globais, fatores demográficos, etc. Esse projeto irá aborar apenas um aspecto geral - o nível de liberdade econômica - e cinco subcomponentes: tamanho do governo, sistema jurídico e direito de propriedade, sound money, liberdade de comércio internacional e regulamentos.

- **Porque o EFW?** Neste projeto, o uso do Índice EFW se justifica devido à sua relevância como uma medida abrangente e amplamente reconhecida para avaliar o nível de liberdade econômica de um país. O índice é baseado em uma estrutura teórica sólida e abrange diversas dimensões, como o tamanho do governo, a eficiência judicial, a liberdade de comércio, sound money e a proteção dos direitos de propriedade. Ao utilizar esse índice, é possível quantificar e comparar o grau de liberdade econômica entre diferentes países, bem como identificar quais subcomponentes têm maior impacto no nível geral de liberdade econômica. A primeira publicação do relatório Economic Freedom of the World foi publicada em 1996, resultado de uma série de conferências realizadas entre 1986 e 1994, com o propósito de desenvolver uma definição clara e quantitativa de liberdade econômica. O índice considera cinco áreas principais, que possuem sub-componentes, que por sua vez também podem apresentar subcomponentes. No total, o índice considera 44 variáveis. Nesse trabalho, iremos trabalhar apenas com o nível geral de liberdade econômica e os 5 grandes componentes do índice, pois o número de subcomponentes variou nas últimas duas décadas e não seria possível obter dados consistentes para todas as variáveis.
- **Componentes do EFW:** O score geral do EFW é resultado da média dos cinco componentes abaixo, que serão utilizados nesse trabalho.

|   | Componentes |
|---|----------|
| 1 | Government Size (Tamanho do governo)        |
| 2 | Legal System & Property Rights (Sistema jurídico e direitos de propriedade)        |
| 3 | Sound money (Moeda sólida)        |                              |
| 4 | Freedom to trade internationally (Liberdade de comércio internacional)        |
| 5 | Regulation (Regulamentos)        |

# 1. Dados do PIB e do índice EFW

Abaixo, carregamos os dados do PIB e filtramos para selecionar o Brasil entre 2000 e 2020, e fazemos o mesmo para o índice EFW. O nome de algumas variáveis foram modificados / encurtados.

```python
# Carregando os dados do PIB (GDP)
url = 'https://raw.githubusercontent.com/bsf94/trabalhoindividual/master/GDPWorldBank.csv'
gdp_data = pd.read_csv(url, skiprows=3)

# Filtrando o dataset do GDP para selecionar apenas o Brasil
brazil_data = gdp_data[gdp_data['Country Name'] == 'Brazil']

# Escolhendo as colunas desejadas (2000-2021)
gdp_columns_to_keep = ['Country Name'] + [str(year) for year in range(2000, 2021)]
brazil_gdp_data = brazil_data[gdp_columns_to_keep]

# Reformulando o formato do dataframe
brazil_gdp_data = brazil_gdp_data.melt(id_vars='Country Name', var_name='Year', value_name='GDP Growth (annual %)')

# Resetando o index
brazil_gdp_data = brazil_gdp_data.reset_index(drop=True)

# Ajuste de display da tabela de GDP
display(HTML("<h2>GDP Data:</h2>"))
display(brazil_gdp_data)

# Especificando os anos para os dados do índice EFW
start_year = 2000
end_year = 2020

# Criação de um dataframe novo para manter apenas as colunas desejadas dos arquivos do índice EFW
result_df = pd.DataFrame(columns=['Year', 'Economic Freedom Summary Index', 'Size of Government', 'Legal System & Property Rights', 'Freedom to trade Internationally', 'Regulation', 'Sound Money'])

# Criando dicionário de variáveis
ef_variable_names = {
    'Economic Freedom Summary Index': 'EFW',
    'Size of Government': 'GovSize',
    'Legal System & Property Rights': 'LegalSystem',
    'Freedom to trade Internationally': 'TradeFreedom',
    'Regulation': 'Regulation',
    'Sound Money': 'SoundMoney'
}

# Loop para executar o código para cada arquivo do índice EFW
for year in range(start_year, end_year + 1):

    # Indicando a URL do arquivo .csv
    ef_url = f'https://raw.githubusercontent.com/bsf94/trabalhoindividual/master/economic_freedom_{year}.csv'

    # Lendo e criando o dataframe dos dados do EFW
    ef_data = requests.get(ef_url).content.decode('utf-8')
    ef_df = pd.read_csv(StringIO(ef_data), skiprows=4)

    # Renomeação das colunas de acordo com os nomes indicados no dicionário
    ef_df = ef_df.rename(columns=ef_variable_names)

    # Anexando os dados do dataframe ao result_df
    result_df = pd.concat([result_df, ef_df], ignore_index=True)

# Resetando o index
result_df = result_df.reset_index(drop=True)

# Mostrando a tabela do EFW
display(HTML("<h2>Economic Freedom Data:</h2>"))
display(result_df[['Year', 'EFW', 'GovSize', 'LegalSystem', 'TradeFreedom', 'Regulation', 'SoundMoney']])
rows, columns = result_df[['Year', 'EFW', 'GovSize', 'LegalSystem', 'TradeFreedom', 'Regulation', 'SoundMoney']].shape
display(HTML(f"<p>Número de linhas: {rows}</p>"))
display(HTML(f"<p>Número de colunas: {columns}</p>"))
```

**Análise inicial dos scores do índice EFW**

* A variável **"EFW"** representa a pontuação geral do índice de liberdade econômica. Quanto maior o valor, maior é a liberdade econômica em um determinado país em relação a outros países. Para o Brasil, os valores variam de 5,77 em 2000 para 6,33 em 2020. Essa medida leva em consideração todas as outras variáveis presentes na tabela para fornecer uma pontuação geral, sendo a média das 5 grandes áreas. A variação entre 2000 e 2020 mostram uma leve melhora nos níveis de liberdade econômica no Brasil, sendo que o “pico” foi em 6,68 em 2010. Esse score é considerado baixo e coloca o Brasil no 3º quartil do índice (de 4 quartis, sendo que países no 1º quartil são considerados os com maior liberdade econômica).

* A variável **"GovSize"** (tamanho do governo) mede o impacto do governo na
economia. Ela inclui elementos como gastos públicos, impostos e regulamentações governamentais. Os dados mostram pouca variação, com valores variando de 6,66 em 2000 para 6,61 em 2020. É importante lembrar que quanto maior o score, melhor é o países em termos de tamanho do governo. De acordo com o liberalismo econômico, interferência do governo na economia não é desejável, logo, quanto menor o score, pior é o papel do governo na economia. O pico de aumentos constantes foi em 2011 com 7,48.

* A variável **"LegalSystem"** (sistema legal) avalia a eficiência e a confiabilidade do sistema legal de um país. Ela leva em consideração aspectos como a proteção dos direitos de propriedade, a independência do poder judiciário e a aplicação da lei. Valores mais altos indicam um sistema legal mais eficiente e confiável. Os dados mostram um aumento gradual dessa variável ao longo dos anos, com valores variando de 4,67 em 2000 para 5,16 em 2020. Ainda assim, é um score muito baixo. O pico foi em 2011 com 5,97.

* A variável **"TradeFreedom"** (liberdade comercial) mede a abertura e a facilidade de realizar transações comerciais em um país. Ela leva em consideração a existência de barreiras comerciais, como tarifas e restrições à importação e exportação. Valores mais altos indicam uma maior liberdade comercial. Os dados mostram uma pequena melhoria, com valores variando de 5,89 em 2000 para 6,17 em 2020. O pico foi em 2006 com 7,33.

* A variável **"Regulation"** (regulamentação) mede o grau de regulamentação governamental na economia, incluindo regulamentações sobre negócios, trabalho e ambiente. Os dados mostram uma variação ao longo dos anos, com valores variando de 5,54 em 2000 para 4,46 em 2020. O pico foi em 2010 com 5,25.

* A variável **"SoundMoney"** (moeda estável) avalia a estabilidade da moeda de um país. Ela leva em consideração aspectos como a inflação e a política monetária. Valores mais altos indicam uma moeda mais estável. Os dados mostram uma variação muito positiva ao longo dos anos, com valores variando de 6,09 em 2000 para 9,25 em 2017.

Podemos observar que Sound Money é a área em que o Brasil é melhor avaliado, enquanto áreas como Legal System e Regulation são preocupantes. Nota-se também que o pico da maioria das variáveis foi entre 2010 e 2011, entretanto, mesmo nos anos de "pico", os scores ainda eram muito baixos, o Brasil nunca saiu do 3º quartil.

### 2.2. Correlacionando os 5 componentes do EFW com o score geral do EFW


Nessa parte, busca-se a correlação entre a variável EFW e os cinco componentes que compõem o índice, para identificar qual dos componentes, no caso do Brasil, tem maior impacto na variação do score geral do nível de liberdade no Brasil conforme medido pelo índice EFW.

```python
# Selecionando as colunas relevantes do dataframe do EFW
ef_components = result_df[['EFW', 'GovSize', 'LegalSystem', 'TradeFreedom', 'Regulation', 'SoundMoney']]

# Calculando a matriz de correlação entre os componentes
correlation_matrix = ef_components.corr()

# Exibindo a matriz de correlação
print("Correlation Matrix:")
display(correlation_matrix)
```

```python
# Gerando uma matrix de correlação
correlation_matrix = ef_components.corr()

# Criando um heatmap para melhor visualização
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='RdYlBu', vmin=-1, vmax=1)
plt.title('Pearson Correlation Matrix')
plt.show()
```

```python
# Calculamos o p-value e indicamos que seja mostrado junto a correlação
def calculate_pvalue(x, y):
    _, p_value = stats.pearsonr(result_df[x], result_df[y])
    return f"{correlation_matrix.loc[x, y]:.3f} ({p_value:.3f})"

# Cópia da matrix de correlação
correlation_matrix_with_pvalues = correlation_matrix.copy()

# Iterando sobre cada par de variável e calculando p-values
for i in range(len(correlation_matrix.columns)):
    for j in range(len(correlation_matrix.columns)):
        if i != j:
            x = correlation_matrix.columns[i]
            y = correlation_matrix.columns[j]
            correlation_matrix_with_pvalues.loc[x, y] = calculate_pvalue(x, y)

# Mostrando a tabela
print("Correlation Matrix with p-values:")
display(correlation_matrix_with_pvalues)
```
